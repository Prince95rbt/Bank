#### Handling Missing values

- Drop all rows/columns with missing values
    - Ideal when we have more rows and less missing values
    - Ideal when all missing values are concentrated in one column
- Fill missing values (missing value imputation)
    - Ideal when he misssing values are low in number. imputation against a column with too many missing values could fudge the aggregations

```python
bank_Data.dropna(axis=1) # Dropping columns with missing values
```

```python
bank_Data.dropna(axis=0) # Dropped all rows with missing values
```

```python
## Filling/Imputation Missing values - Central Imputation
```

```python
bank_Data.Age = bank_Data.Age.fillna(value = bank_Data.Age.mean()) # Filling age missing values with mean
bank_Data.Age = bank_Data.Age.fillna(value = bank_Data.Age.median()) # Filling age missing values with median
bank_Data['Education'] = bank_Data['Education'].fillna(value=bank_Data.Education.mode()) # Filling missing values with most frequent observation
```
